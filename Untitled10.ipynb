```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler, LabelEncoder
from sklearn.linear_model import LogisticRegression
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import classification_report
import matplotlib.pyplot as plt
import numpy as np

# Load the uploaded dataset
df = pd.read_csv('https://raw.githubusercontent.com/hngahsb/churn-analysis/refs/heads/main/WA_Fn-UseC_-Telco-Customer-Churn.csv')
df.head()
# Convert 'TotalCharges' to numeric, handle non-numeric values, and drop NA rows
df['TotalCharges'] = pd.to_numeric(df['TotalCharges'], errors='coerce')
df = df.dropna()

# Drop 'customerID' column
df = df.drop(columns=['customerID'])

# Encode categorical variables
categorical_columns = df.select_dtypes(include=['object']).columns
label_encoders = {col: LabelEncoder() for col in categorical_columns}
for col in categorical_columns:
    df[col] = label_encoders[col].fit_transform(df[col])

# Define features and target
X = df.drop(columns='Churn')
y = df['Churn']

# Split the data
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

# Standardize the features for KNN
scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)

# Train Logistic Regression model
log_reg = LogisticRegression(random_state=42, max_iter=1000)
log_reg.fit(X_train, y_train)
y_pred_log_reg = log_reg.predict(X_test)

# Train KNN model
knn = KNeighborsClassifier(n_neighbors=5)
knn.fit(X_train_scaled, y_train)
y_pred_knn = knn.predict(X_test_scaled)

# Generate classification reports
log_reg_report_dict = classification_report(y_test, y_pred_log_reg, output_dict=True)
knn_report_dict = classification_report(y_test, y_pred_knn, output_dict=True)

# Plotting precision, recall, and f1-score for both models
metrics = ['precision', 'recall', 'f1-score']
classes = ['0', '1']

fig, axs = plt.subplots(1, 3, figsize=(18, 6))
fig.suptitle("Logistic Regression vs KNN Classification Report Metrics")

for i, metric in enumerate(metrics):
    log_reg_values = [log_reg_report_dict[cls][metric] for cls in classes]
    knn_values = [knn_report_dict[cls][metric] for cls in classes]

    # Set width and positions for bars
    x = np.arange(len(classes))
    width = 0.35

    # Plot bars
    axs[i].bar(x - width/2, log_reg_values, width, label='Logistic Regression')
    axs[i].bar(x + width/2, knn_values, width, label='KNN')

    # Labeling and adjustments
    axs[i].set_title(f"{metric.capitalize()} Comparison")
    axs[i].set_xlabel("Class")
    axs[i].set_ylabel(metric.capitalize())
    axs[i].set_xticks(x)
    axs[i].set_xticklabels(classes)
    axs[i].legend()

plt.tight_layout(rect=[0, 0.03, 1, 0.95])
plt.show()
```

```python
from sklearn.model_selection import cross_val_score
from sklearn import neighbors

k_values = range(1, 21)
accuracy_scores = []

# Experiment with k values from 1 to 20
for k in k_values:
    knn = KNeighborsClassifier(n_neighbors=k)
    scores = cross_val_score(knn, X_train_scaled, y_train, cv=5, scoring='accuracy')
    accuracy_scores.append(scores.mean())

# Find the k value with the highest accuracy
optimal_k = k_values[np.argmax(accuracy_scores)]
print("The optimal k value is:", optimal_k)
```

```python


# Huấn luyện mô hình hồi quy tuyến tính
linreg = LinearRegression()
linreg.fit(X_train, y_train)

# Dự đoán xác suất churn
y_pred_prob = linreg.predict(X_test)

# Làm tròn xác suất để phân loại churn
y_pred_linreg = [1 if x > 0.5 else 0 for x in y_pred_prob]

print("Độ chính xác Linear Regression:", accuracy_score(y_test, y_pred_linreg))
print(classification_report(y_test, y_pred_linreg))

# Biểu đồ ma trận nhầm lẫn
sns.heatmap(confusion_matrix(y_test, y_pred_linreg), annot=True, fmt='d', cmap='Greens')
plt.title("Confusion Matrix - Linear Regression")
plt.show()

```

```python
# Huấn luyện mô hình Logistic Regression
logreg = LogisticRegression()
logreg.fit(X_train, y_train)

# Dự đoán và đánh giá
y_pred_logreg = logreg.predict(X_test)
print("Độ chính xác Logistic Regression:", accuracy_score(y_test, y_pred_logreg))
print(classification_report(y_test, y_pred_logreg))

# Biểu đồ ma trận nhầm lẫn
sns.heatmap(confusion_matrix(y_test, y_pred_logreg), annot=True, fmt='d', cmap='Blues')
plt.title("Confusion Matrix - Logistic Regression")
plt.show()
```
